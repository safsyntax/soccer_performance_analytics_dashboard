# Soccer Data Analysis

```
import pandas as pd

# Load
results = pd.read_csv("results_cleaned.csv")
goalscorers = pd.read_csv("goalscorers_cleaned.csv")
shootouts = pd.read_csv("shootouts_cleaned.csv")

print("Results shape:", results.shape)
print("Goalscorers shape:", goalscorers.shape)
print("Shootouts shape:", shootouts.shape)

# Match Dist. Over Time
results['date'] = pd.to_datetime(results['date'])
results['year'] = results['date'].dt.year

matches_per_year = results.groupby('year').size().reset_index(name='total_matches')

print("\n=== Matches Over Time ===")
print(matches_per_year)

import matplotlib.pyplot as plt
matches_per_year.plot(x='year', y='total_matches', kind='line', title='Matches Over Time')
plt.show()

# Match Outcome Dist
results['home_win'] = (results['home_score'] > results['away_score']).astype(int)
results['away_win'] = (results['home_score'] < results['away_score']).astype(int)
results['draw'] = (results['home_score'] == results['away_score']).astype(int)

outcomes = results[['home_win', 'away_win', 'draw']].sum()

print("\n=== Match Outcomes ===")
print(outcomes)

# Average Goals
results['total_goals'] = results['home_score'] + results['away_score']
results['goal_diff'] = abs(results['home_score'] - results['away_score'])

avg_goals = results[['total_goals', 'goal_diff']].mean()

print("\n=== Goal Statistics ===")
print(avg_goals)

# Top Teams by Matches
home_counts = results['home_team'].value_counts()
away_counts = results['away_team'].value_counts()

total_matches = home_counts.add(away_counts, fill_value=0).reset_index()
total_matches.columns = ['team', 'matches_played']

top_teams = total_matches.sort_values(by='matches_played', ascending=False).head(10)

print("\n=== Top Teams by Matches Played ===")
print(top_teams)

# Home Win Rate
home_games = results.groupby('home_team').agg(
    home_games=('home_team', 'count'),
    home_wins=('home_win', 'sum')
).reset_index()

home_games['win_rate'] = home_games['home_wins'] / home_games['home_games']

home_games = home_games[home_games['home_games'] > 20]
home_games = home_games.sort_values(by='win_rate', ascending=False)

print("\n=== Home Win Rate by Team ===")
print(home_games.head(10))

# Matches by Tournament
tournament_counts = results['tournament'].value_counts().reset_index()
tournament_counts.columns = ['tournament', 'total_matches']

print("\n=== Matches by Tournament ===")
print(tournament_counts.head(10))

# Matches by Country
country_counts = results['country'].value_counts().reset_index()
country_counts.columns = ['country', 'total_matches']

print("\n=== Top Countries by Matches ===")
print(country_counts.head(10))
```
